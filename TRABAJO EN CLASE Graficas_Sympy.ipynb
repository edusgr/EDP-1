### Graficas 3D

```python
from sympy.plotting import plot3d
```

permite mostrar graficos en el espacio

```python
x,y=symbols('x y')
```

```python
plot3d(x*y)
```

muestra la funcion $z=xy$ en el espacio

```python
plot3d(x*y, (x,-5,5),(y,-5,5))
```

de la misma funcion restringe el dominio para cada entrada de la funcion $f(x,y)$

```python
plot3d(x*y, -x*y, (x,-5,5),(y,-5,5))
```

se pueden agregar mas funciones al mismo lienzo, separadas por comas, y restringir los dominios de estas variables

```python
plot3d((x**2 + y**2, (x,-5,5),(y,-5,5)),(x*y,(x,-3,3),(y,-3,3)))
```

esta es otra manera de poner los graficos en el mismo lienzo, pero cada una de estas con sus restricciones distintas, esto sera la funcion con su restriccion y separando con coma, la otra funcion y su restriccion

```python
from sympy.plotting import plot3d_parametric_line
```

permite escribir ecuaiones parametricas que formen lineas, no solo superficies

```python
plot3d_parametric_line(cos(s),sin(s),s)
```

se define una ecuacion para x en la primera coordenada del comando y pàra y en la segunda entrada del arreglo, que resultan en lineas

```python
plot3d_parametric_line((cos(s),sin(s),s,(s,-5,5)),(sin(s),s**2,s,(s,-5,5)))
```

del mismo modo que las anteriores se restringen las ecuacines utiliznado comas

```python
from sympy.plotting import plot3d_parametric_surface
```

se utiliza para graficar superficies paramétricas en 3D. Una superficie paramétrica en 3D está definida por tres funciones, una para cada una de las coordenadas $x,y,z$ en términos de dos parámetros.

```python
u, v =symbols('u v')
```

estos son los parametros

```python
plot3d_parametric_surface(cos(u+v),sin(u-v),u-v)
```

```python
plot3d_parametric_surface(cos(v)*cos(u),cos(v)*sin(u),sin(v))
```

```python
from sympy import plot_implicit, symbols, Eq, And
```

permiten graficar ecuaciones implícitas y trabajar con expresiones lógicas.

```python
x, y = symbols(' x y')
```

```python
p1 = plot_implicit(Eq(x**2 + y**2, 5))
```
