INCISO c)

Comprobación algebraica de $(1-x^2)P'_n(x)=(n+1)xP_n(x)-(n+1)P_{n+1}(x)$

con $n=3$

```python
from sympy import symbols, legendre, diff, simplify, latex
from IPython.display import display, Math

x = symbols('x')
n = 3
Pn_expr = legendre(n, x)
Pn1_expr = legendre(n + 1, x)
dPn_dx = diff(Pn_expr, x)

lhs = (1 - x**2) * dPn_dx
rhs = (n + 1) * x * Pn_expr - (n + 1) * Pn1_expr

# Mostrar las expresiones en LaTeX
display(Math(r"\textbf{Lado izquierdo (I):} \quad (1 - x^2) \cdot \frac{d}{dx}P_3(x) = " + latex(lhs)))
display(Math(r"\textbf{Lado derecho (D):} \quad 4x \cdot P_3(x) - 4 \cdot P_4(x) = " + latex(rhs)))

# Simplificación
simpl_diff = simplify(lhs - rhs)
display(Math(r"\textbf{Simplificación de I - D=} \quad " + latex(simpl_diff)))

```

INCISO d)

Comprobación algebraica de $(n+1)P_{n+1}(x)-(2n+1)xP_n(x)+nP_{n-1}(x)=0$

con $n=0,1,2,...$

```python
from sympy import symbols, legendre, diff, simplify, latex
from IPython.display import display, Math

x = symbols('x')
n = 3

#polinomios de Legendre
Pn_2 = legendre(n - 1, x)
Pn_3 = legendre(n, x)
Pn_4 = legendre(n + 1, x)

#expresion
expr = (n + 1) * Pn_4 - (2 * n + 1) * x * Pn_3 + n * Pn_2

#mostrar sustitucion
display(Math(r"\textbf{Sustituyendo } n=3:"))
display(Math(latex(expr)))

#simplificar
simpl_expr = simplify(expr)
display(Math(r"\textbf{Simplificando=} \quad " + latex(simpl_expr)))
```
