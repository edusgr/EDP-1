```python
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

importamos las librerias con las que trabajaremos para simplificar las cosas

```python
def f1(x):                    # primer trozo de la funcion f(x)=0
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 0
    return y
def f2(x):                    # segundo trozo f(x)= 2-x
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 2 - x[i]
    return y
```

Definimos la funci√≤n a trozos como dos funciones independientes, es decir, graficamos cada una por separado

```python
x1 = np.linspace(-10,0,10000)
x2 = np.linspace(0,20,10000)
y1 = f1(x1)
y2 = f2(x2)
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$ x + 2$ para $-2 \leq x \leq 0$')
plt.plot(x2, y2, label=r'$ 0$ para $0 x \leq x \leq 2$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("$f(x)$")
xlim([-10.01,10.01]),ylim([-7.01,10.01]),grid(),legend()
show()
```

```python
def u_0(x,N=40): # n√πmero de armocios para la aproximacion
    f = zeros(len(x))
    for n in range(1,N):
        f += ((2-2*(-1)**n)/(np.pi**2 * n**2) * np.cos((n*np.pi*x)/2) + (2/(np.pi*n)) * np.sin((n*np.pi*x)/2)) # += ES UNA SUMATORIA, este f es la serie de Fourier
    return (1/2)+f
```

Definimos la serie de fourier para nuestra funci√≤n respectivamente, en este caso, consideramos un solo $f(x)$

```python
x = np.concatenate((x1, x2))
f_n = u_0(x)
plot(x1, y1, label='$f(x) = 0$ para $-2 \leq x \leq 0$')
plot(x2, y2, label='$f(x) = 2-x$ para $0 x \leq x \leq 2$')
plot(x, f_n, label='$f_N(x)$ con $N=20$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci√≥n de la condici√≥n inicial")
xlim([-2.0,2.1]),ylim([-0.3,2.3]),grid(),legend()
show()
```

Observamos la serie para 20 armonicos, este parametro se puede modificar si queremos una mejor aproximaci√≤n

```python
for i in range (1,21,5):  # el intervalo de armonicos que nos interesa analizar, y el paso
  plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
  plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci√≥n con arm√≤nicos")
xlim([-2.0,2.1]),ylim([-0.3,2.3]),grid(),legend()
show()
```

Observamos como se puede comportar las aproximaciones a medida que aumentamos los armonicos en un determinado intervalo.

Por el teorema, vemos que f es suave a trozos en [-2,2], entonces la serie converge puntualmente a la funci√≤n
