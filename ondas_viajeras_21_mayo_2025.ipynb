```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
t = 1
x = np.linspace(-2, 2.5, 500)

def f(xp):
 return np.exp(-((xp - 0.5)**2) / 0.01)

f_derecha = f(x - c*t)
f_izquierda = f(x + c*t)
f_superpuesta = f_derecha + f_izquierda

```

GRAFICA DE LA FUNCION

$f(x')=e^{-\frac{(x'-0.5)^2}{0.01}}$

considerando los siguientes cambios de variable

$x'=x-ct \hspace{0.5 cm}$ indica que se desplaza a la derecha $\hspace{0.5 cm}(Fig. 1.)$

$x'=x+ct \hspace{0.5 cm}$ indica que se desplaza a la izquierda $\hspace{0.5 cm}(Fig. 2.)$

Y se grafica la superposición de ambas gráficas $\hspace{0.5 cm}(Fig. 3.)$

```python
# graficas
plt.figure(figsize=(12, 6))

# a la derecha
plt.subplot(1, 3, 1)
plt.plot(x, f_derecha, label='derecha', color='purple')
plt.title('desplazada a la derecha')
plt.xlabel('Fig. 1')
plt.grid()

# a la izquierda
plt.subplot(1, 3, 2)
plt.plot(x, f_izquierda, label='izquierda', color='red')
plt.title('desplazada a la izquierda')
plt.xlabel('Fig. 2')
plt.grid()

# super posicion
plt.subplot(1, 3, 3)
plt.plot(x, f_superpuesta, label='superposición', color='blue')
plt.title('superposicion')
plt.xlabel('Fig. 3')
plt.grid()

plt.tight_layout()
plt.show()
```
