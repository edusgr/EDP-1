```python
from sympy import fourier_series, pi, plot, Piecewise, E
from sympy.abc import x
```

```python
f=Piecewise(
    (E**(2*x), (x > 0) & (x <= 1)),
    (E**(-2*x), (x >= -1) & (x <= 0))
)
```

```python
plot(f, (x,-2,2),line_color ='blue')
```

```python
ss = fourier_series(f, (x, -1, 1))
ss1 = ss.truncate(n = 3)
ss2 = ss.truncate(n = 5)
ss3 = ss.truncate(n = 7)
ps = plot(f, ss1, ss2, ss3, (x, -1, 1), show=False, legend=True)
```

```python
ss
```

```python
ps[0].label = 'x'
ps[1].label = 'n=3'
ps[2].label = 'n=5'
ps[3].label = 'n=7'
```

```python
ps.show()
```

AHORA SE CALCULA PARA LA EXTENSION IMPAR

```python
Y=Piecewise(
    (E**(2*x), (x > 0) & (x <= 1)),
    ((-1)*E**(-2*x), (x >= -1) & (x <= 0))
)
```

```python
plot(Y, (x,-1,1),line_color ='green') #FUNCION IMPAR
```

```python
s = fourier_series(Y, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(Y, s1, s2, s3, (x, -1, 1), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

FOURIER DE UNA FUNCIÒN A TROZOS

```python
W=Piecewise(
    (1, (x >= 0) & (x <= pi/2)),
    (2, (x > pi/2) & (x <= pi))
)
```

```python
plot(W, (x,-2,pi),line_color ='red')
```

```python
st = fourier_series(W, (x, 0, pi))
st1 = st.truncate(n = 3)
st2 = st.truncate(n = 5)
st3 = st.truncate(n = 7)
pt = plot(W, st1, st2, st3, (x, 0, pi), show=False, legend=True)
```

```python
st
```

```python
pt[0].label = 'x'
pt[1].label = 'n=3'
pt[2].label = 'n=5'
pt[3].label = 'n=7'
```

```python
pt.show()
```
