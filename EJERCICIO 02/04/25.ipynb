Tenemos la siguiente función:
$$X''-λX=0$$
con las condiciones
$$  X(0)=X(1)=0$$

Y tenemos la siguiente función:
$$X''-λX=0$$
con las condiciones
$$  X'(0)=X'(1)=0$$


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Variables simbólicas
x, lam = sp.symbols('x lambda', real=True)
X = sp.Function('X')

# Soluciones simbólicas
n = sp.symbols('n', integer=True, positive=True)
lambda_n = (n * sp.pi)**2

# Funciones propias
Xn_dirichlet = sp.sin(n * sp.pi * x)
Xn_neumann = sp.cos(n * sp.pi * x)

# Lambdify para graficar (crear funciones evaluables numéricamente)
Xn_dirichlet_fns = [sp.lambdify(x, Xn_dirichlet.subs(n, i), 'numpy') for i in range(1, 4)]
Xn_neumann_fns  = [sp.lambdify(x, Xn_neumann.subs(n, i), 'numpy') for i in range(1, 4)]

# Valores de x para graficar
x_vals = np.linspace(0, 1, 200)

# Graficar funciones propias (Dirichlet)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_dirichlet_fns, 1):
    plt.plot(x_vals, f(x_vals))
plt.title("X''-λX=0  con las condiciones  X(0)=X(1)=0")
plt.xlabel("x")
plt.ylabel("X(x)")

plt.grid(True)
plt.show()

print()

# Graficar funciones propias (Neumann)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_neumann_fns, 1):
    plt.plot(x_vals, f(x_vals))
plt.title("X''-λX=0  con las condiciones  X'(0)=X'(1)=0")
plt.xlabel("x")
plt.ylabel("X(x)")

plt.grid(True)
plt.show()


print("Soluciones con x(0)=x(1)=0:")
for i in range(1,4):
  print(f"n={i}:X_n(x)=",Xn_dirichlet.subs(n,i))
  print("Soluciones con x'(0)=x'(1)=0:")
for i in range(1,4):
  print(f"n={i}:X_n(x)=",Xn_neumann.subs(n,i))
```
